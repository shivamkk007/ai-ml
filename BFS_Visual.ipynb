```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
node_pos = {'A':(5,0),'B':(4,-1),'C':(6,-1),'D':(3,-2),'E':(5,-2),'F':(6,-3)}
```

```python
G = nx.Graph()

G.add_nodes_from(['A','B','C','D','E','F'])
G.add_edges_from([('A','B'),('A','C'),('B','E'),('B','D'),('E','F')])

nx.draw_networkx(G,node_pos,font_color = 'white', node_shape = 's')

G_dict = {node: list(G[node]) for node in G}

visited = []                                                # List for visited nodes.
queue = []                                                  #Initialize a queue

def bfs(visited, G_dict, node):                              #function for BFS
    visited.append(node)
    queue.append(node)
    while queue:                                              # Creating loop to visit each node
        m = queue.pop(0) 
        print (m, end = " ") 
        for neighbour in G_dict[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

# Driver Code
print("Breadth-First Search")
bfs(visited, G_dict, 'A')  
```
